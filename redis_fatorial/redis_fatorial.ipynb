# Importar Bibliotecas

```python
import redis
from memoizit import Memoizer

m = Memoizer(
    backend="redis",
    host="localhost",
    port="6379"    
)
```

# Conexão com redis

```python
redis_connection = redis.Redis(host='localhost', port=6379, decode_responses=True)
```

# Criar métodos

```python
def fatorial(n):
    if n < 0:
        return "Não é possível calcular o fatorial de um número negativo."
    elif n == 0:
        return 1
    else:
        return n * fatorial(n - 1)

def cached_fatorial(n : int) -> int:
    k = f"fatorial:{n}"
    cf = redis_connection.get(k)
    if cf is None:
        cf = fatorial(n)
        print("Vai colocar no cache -> ", k, cf)
        redis_connection.set(k, cf)
    else:
        print("Usando cache -> ", k, cf)
    return cf
```

```python

# Exemplo de uso:
for numero in range(1, 100):
    print(f'O fatorial de {numero} é {cached_fatorial(numero)}')
```

# Utilizar Memoizit

```python
@m.memoize(
    expiration=60*60*24
)
def memo_fatorial(n):
    if n < 0:
        return "Não é possível calcular o fatorial de um número negativo."
    elif n == 0:
        return 1
    else:
        return n * memo_fatorial(n - 1)
```

```python
# Exemplo de uso:
for numero in range(1, 100):
    print(f'O fatorial de {numero} é {memo_fatorial(numero)}')
```
