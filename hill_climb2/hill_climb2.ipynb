```python
import random
import matplotlib.pyplot as plt

# Função para calcular f(x)
def f(x):
    return -x**2 + 4*x

# Função para encontrar o máximo local usando Hill Climbing
def hill_climbing():
    # Lista para armazenar os pontos explorados
    explored_points = []

    # Inicialização: Escolha um valor inicial aleatório para x no intervalo [0, 4]
    x = random.uniform(0, 4)
    explored_points.append((x, f(x)))

    # Definição do valor de delta para a vizinhança
    delta = 0.01

    # Número máximo de iterações para evitar loops infinitos
    max_iterations = 1000

    # Iteração
    for _ in range(max_iterations):
        # Avaliação: Calcule f(x) para o ponto atual e seus vizinhos
        current_score = f(x)
        left_neighbor_score = f(x - delta)
        right_neighbor_score = f(x + delta)

        # Verifique se os vizinhos têm valores maiores de f(x) e atualize x se sim
        if left_neighbor_score > current_score:
            x -= delta
            explored_points.append((x, f(x)))
        elif right_neighbor_score > current_score:
            x += delta
            explored_points.append((x, f(x)))
        else:
            # Critério de parada: Não há melhoria possível
            break

    return explored_points

# Executar o algoritmo de Hill Climbing
explored_points = hill_climbing()

# Plotar a função e os pontos explorados
x_values = [point[0] for point in explored_points]
y_values = [point[1] for point in explored_points]

plt.plot(x_values, y_values, marker='o', color='red', label='Pontos explorados')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Algoritmo de Hill Climbing')
plt.grid(True)
plt.legend()
plt.show()
```
